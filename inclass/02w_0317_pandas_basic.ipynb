## 시리즈 활용

```python
import numpy as np
import pandas as pd

s = {'희빈', '수영', '지민', '유정'}
p = pd.Series([99,48,64,84], index=s)

p
```

```python
p.name = '수학성적'
p.index.name = '이름'
p
```

```python
p.sort_index(ascending=False) #오름차순
```

```python
p.sort_values()
```

```python
p.sort_values(ascending=False) #오름차순
```

## 시리즈 기본 통계 함수

```python
print(p.max())
print(p.min())
print(p.median())
print(p.mean())
print(p.count())
print(p.size)
```
