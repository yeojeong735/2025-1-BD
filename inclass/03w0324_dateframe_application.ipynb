# 데이터프레임 수정

```python
df[['사망(명)','부상(명)']].sum(axis=1)
```

```python
df.iloc[:, 1:].sum(axis=1)
```

```python
df2 = df.copy()
df2['재해자수'] = df[['사망(명)','부상(명)']].sum(axis=1)
df2
```

```python
df2.loc['총계'] = df2.sum()
df2
```

```python
df2 > 10
```

```python
df2[df2>10]
```

```python
df2[[True,False]*6 + [True]]
```

```python
df2[[True]*6 + [False]*7]
```

```python
df2['재해자수']>400
```

```python
df2[df2['재해자수']>400]
```

```python
df['사망(명)'] == 7
```

```python
df2[ df2['사망(명)'] == 7]
```

```python
df2[ df2['사망(명)'].isin([7,4])]
```

```python
df2[(df2['사망(명)'].isin([7,10])) | (df2['부상(명)']>350)]
```

```python
df2[df2.index.str.contains('3')]
```
