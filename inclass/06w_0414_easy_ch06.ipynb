## assign( ) 메서드 : DataFrame에서 새로운 열을 추가하거나 기존 열을 수정할 때 사용하는 메서드 (원본 수정 X)

## 람다 함수 : 이름 없이 한 줄로 정의하는 간단한 함수
### -> (lambda 매개변수 : 반환값)

```python
# 람다 함수
(lambda x, y: x + y)(30, 40)
```

```python
f = lambda x, y: x + y
f(30, 4)
```

```python
exam.assign(total = exam.math + exam.english + exam.science,
            # lambda x: x.total / 3는 .assign()이 total을 추가한 후, 내부적으로 계산되기 때문에 에러 없이 실행됨.
            mean = lambda x: x.total / 3) # = x['total']
```

```python
round(3.4534,3)
```

```python
exam.assign(total = exam.math + exam.english + exam.science,
            mean = lambda x: round(x.total / 3, 3)) # total 값을 3으로 나누고 결과를 소수점 셋째 자리까지 반올림
```

```python
pd.set_option('display.precision', 2)
# display.precision : Pandas가 DataFrame이나 Series를 출력할 때 소수점을 몇 자리까지 보여줄지 지정
```

```python
exam.assign(total = exam.math + exam.english + exam.science,
            mean = lambda x: x.total / 3)
```

```python
exam.assign(total = lambda x: x.math + x.english + x.science,
            mean = lambda x: x.total / 3)
```

## groupby( ) : DataFrame에서 특정 열의 값들을 기준으로 묶어서 그룹별 집계를 할 수 있게 해주는 함수

```python
exam.groupby('nclass').agg(mean_math = ('math','mean'))
# ㄴ 'nclass'라는 열의 값을 기준으로 같은 값들끼리 묶는다는 뜻
```

```python
exam.groupby('nclass').agg('mean')
```

```python
exam.groupby('nclass').mean()
```

```python
exam.groupby('nclass').mean()[["math"]]
```

```python
exam.groupby('nclass').mean()[["math","science"]]
```

```python
exam.groupby('nclass').agg('mean')['math']
```

```python
df1 = exam.groupby('nclass').agg(['mean', 'median'])
```

```python
df1
```

```python
df1.columns
```

```python
exam.groupby('nclass').agg(mean_math = ('math','mean'),
                           mean_eng = ('english','mean'))
```

```python
exam.groupby('nclass').agg('mean')['math']
# 반환값 형태 : Series
# 모든 열 평균 후 math만 선택
```

```python
exam.groupby('nclass').agg({'math' : 'mean'})
# 반환값 형태 : DataFrame
# math만 집
```

```python
exam.groupby('nclass').agg({'math' : 'mean','english': 'mean'})
```

```python
exam.groupby('nclass').agg({'math':['mean','min','max']})
```

```python
exam.groupby('nclass').agg({'math':['mean','min','max'],'science':['mean','min','max']})
```

```python
exam.groupby('nclass').agg(mean_math = ('math','mean'))
```

```python
exam.groupby('nclass', as_index=True).agg(mean_math = ('math', 'mean'))
# as_index=True : 그룹 기준 컬럼을 인덱스로
```

```python
exam.groupby('nclass', as_index=False).agg(mean_math = ('math', 'mean'))
# as_index=False : 그룹 기준 컬럼을 일반 열로 유지
```

## mpg 데이터 활용 groupby() agg()

```python
mpg = pd.read_csv('data/mpg.csv')
mpg.info()
```

```python
mpg.columns
```

```python
mpg.sample(frac=0.2)
```

```python
mpg.groupby(['manufacturer','drv']).agg(mean_cty=('cty','mean'))
```

```python
mpg.groupby(['manufacturer', 'category']).agg(mean_cty = ('cty', 'mean'), mean_hwy = ('hwy', 'mean'))
```

```python
mpg.groupby(['category', 'manufacturer']).agg(mean_cty = ('cty', 'mean'), mean_hwy = ('hwy', 'mean'))
```

```python
mpg.query('manufacturer == "audi"').groupby(['drv']).agg(n = ('drv', 'count'))
# 아우디 차량만 필터링 -> drv 별로 나눔 -> drv 열의 값을 count
```

```python
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).agg(n = ('drv', 'count'))
```

```python
mpg.query('manufacturer == "chevrolet"') \
   .groupby(['drv']) \
   .agg(n = ('drv', 'count'))
```
