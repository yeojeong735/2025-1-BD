## 시리즈 활용

```python
import numpy as np
import pandas as pd

s = {'희빈', '수영', '지민', '유정'}
p = pd.Series([99,48,64,84], index=s)

p
```

```python
p.name = '수학성적'
p.index.name = '이름'
p
```

```python
p.sort_index(ascending=False) #index 내림차순 / ascending = 상승
```

```python
p.sort_values() # DataFrame.sort_values(by='정렬의 기준이 될 이름', ascending=True, inplace=False) :데이터프레임에서 특정 열을 기준으로 정렬할 때 사용
                # - inplace : 원본 데이터 변경 여부
```

```python
p.sort_values(ascending=False) #오름차순
```

## 시리즈 기본 통계 함수

```python
print(p.max())
print(p.min())
print(p.median())
print(p.mean())
print(p.count())
print(p.size)
```

```python
p.describe() # 다양한 통계값 일괄 표시
```
