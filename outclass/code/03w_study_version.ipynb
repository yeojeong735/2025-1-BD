```python
df[df.A > df.B]
```

```python
df.query('C == 10')
```

```python
df.query('A>0')
```

```python
df.query('E == "python"')
```

```python
df.query("E != 'python'")
```

```python
df.query('A == `D clmns`')
```

```python
df.query('A == "D clmns"')
```
